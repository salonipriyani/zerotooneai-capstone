```python
import os
import requests
import requests
from bs4 import BeautifulSoup
```

```python
import csv

input_folder = 'review_folder'
rest_list = []
for input_file in os.listdir(input_folder):
    if input_file.endswith(".csv"):
        rest_name = input_file.split('.csv')[0]
        rest_list.append(rest_name)
rest_list.append("union grill")
rest_list.append("El Sabor Latin Kitchen")
rest_list.append("The Forge")
rest_list.append("Butterjoint")
rest_list.append("The Columbian Spot")
rest_list.append("DiAnoia's Eatery")
rest_list.append("Raising Canes")
rest_list.append("The Porch at Schenley")
rest_list.append("Spirits & Tales")
rest_list.append("Gaucho Parrilla Argentina")
```

```python
output_folder = 'raw_text_files'
for rest_name in rest_list:
    print(rest_name)
    raw_text = ''
    raw_text += f"Information about {rest_name} Restaurant"
    url = "https://api.yelp.com/v3/businesses/search"

    headers = {
        "accept": "application/json",
        "Authorization": "Bearer vc35EpBzhOzx4k6_0FM4rNgcB8cHtZdzECD-wYho9WFB9FM3iklf4ud5kvfXU2AU5g7Xna2p1o6MV_kq4oR-hQPE_Uu8R1ntPVkf6yVFPOKZPeGqnu0-KXyCMJwwZXYx"
    }
    parameters = {
        "location" : "Oakland, Pittsburgh, PA",
        "limit": 1,
        "term": rest_name
    }
    response = requests.get(url, headers=headers, params=parameters)
    raw_text += response.text
    yelp_url = response.json()['businesses'][0]['url']
    print(yelp_url)
    response = requests.get(yelp_url)
    soup = BeautifulSoup(response.content, "html.parser")
    dishes = soup.find_all("p", class_="css-nyjpex")
    raw_text += 'Famous dishes of this restaurant include:'
    for dish in dishes:
        print(dish.text.strip())
        raw_text += (dish.text.strip() + ',')

    raw_text += "\n"
    review_dict_list = []
    current_page = 1
    total_reviews = 0
    base_url = yelp_url.split("?")[0]
    print(base_url)
    while total_reviews < 100:
        if current_page != 1:
            url = f'{base_url}?start={(current_page - 1) * 10}'
        else:
            url = base_url
        print(url)
        r = requests.get(url).text
        soup = BeautifulSoup(r,'html.parser')
        review_elements = soup.find_all("li", class_="css-1q2nwpv")
        for review in review_elements:
            review_1 = {}
            review_text = review.find('span', class_='raw__09f24__T4Ezm')
            review_rating = review.find('div', class_='five-stars__09f24__mBKym five-stars--regular__09f24__DgBNj css-1jq1ouh')
            if review_rating is not None:
                review_rating = review_rating.get('aria-label')
            if review_text:
                review_1["review_text"] = review_text.text
                review_1["review_rating"] = review_rating
                review_dict_list.append(review_1)
                total_reviews += 1
        next_button = soup.find('a', class_='next-link')
        if next_button:
            current_page += 1
        else:
            break
    raw_text += str(review_dict_list)

    with open(os.path.join(output_folder,f'{rest_name}.txt'), 'w') as f:
        f.write(raw_text)
        
        
```
